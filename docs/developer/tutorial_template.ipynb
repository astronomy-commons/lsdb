## Open a catalog

```python
import lsdb
```

```python
# At bottom of imports ->  Dask emits a lot of `INFO` logging. Raise the threshold to `WARNING`
# so tutorial output stays readable
import logging

logging.getLogger().setLevel(logging.WARNING)  # LSST / Butler / general INFO
logging.getLogger("distributed").setLevel(logging.WARNING)  # Dask scheduler / worker INFO
```

```python
# print out DASK dashboard link explicitly
from dask.distributed import Client

client = Client(n_workers=4, threads_per_worker=1, memory_limit="auto")
print(f"Dask dashboard: {client.dashboard_link}")
client
```

```python
ztf_object_path = "https://data.lsdb.io/hats/ztf_dr22/ztf_lc"
ztf_object = lsdb.open_catalog(ztf_object_path)
ztf_object
```

## Plotting lightcurves

Standard recipe for a single-object lightcurve: `errorbar` of magnitude vs. MJD with an inverted y-axis. Adapt the column names to your catalog.

```python
import matplotlib.pyplot as plt

ztf_nested = ztf_object.nest_lists(list_columns=["hmjd", "mag", "magerr"], name="lc")
lc = ztf_nested.head(1).iloc[0]["lc"]

plt.errorbar(lc["hmjd"], lc["mag"], lc["magerr"], fmt="o", markersize=3, alpha=0.6, ls="none")
plt.gca().invert_yaxis()
plt.xlabel("MJD")
plt.ylabel("Magnitude")
plt.show()
```
