## Introduction

### Prerequisites

In order to access Rubin data, you must be a [Rubin data rights holder](https://rubinobservatory.org/for-scientists/data-products/data-policy).

At present, there are two ways to access this data: RSP and CANFAR, described [here](./rubin_dp1.html).

**NOTE:** Any Rubin data rights holder can access data from any Independent Data Access Center (IDAC),
or Rubin Science Platform (RSP).  (CANFAR is an example of a platform served by a Canadian IDAC.)

## 1. Cross-match Rubin DP1 and VSX

We load both catalogs with LSDB.
You may need to change Rubin DP1 path, while HATS version of VSX is hosted by https://data.lsb.io

```python
import lsdb
from dask.distributed import Client
from upath import UPath
```

### 1.1 Load the catalog metadata

```python
# Change it to the actual location, here we use USDF path.
DP1_PATH = UPath("/sdf/data/rubin/shared/lsdb_commissioning/dp1/dp1_release/")

dp1_objects = lsdb.open_catalog(DP1_PATH / "object_collection", columns="all")
dp1_objects
```

```python
vsx = lsdb.open_catalog("https://data.lsdb.io/hats/vsx_2025-03-21/vsx/")
vsx
```

### 1.2 Select eclipsing binaries

```python
eclipsing = vsx.query("'EA' in Type or 'EB' in Type or 'EW' in Type")
```

### 1.2 Plan the cross-match

Since LSDB loads and plans operations lazily (see [this doc page](./lazy_operations.html)), no data has been loaded yet, and the next cell does not run the cross-match pipeline either.

```python
dp1_eclipsing = eclipsing.crossmatch(dp1_objects, radius_arcsec=0.1, suffixes=["_vsx", "_rubin"])
dp1_eclipsing
```

### 1.4 Run the pipeline

Now it time to create a local Dask cluster and actually run the pipeline. See [this doc page](./dask-cluster-tips.html) for additional information about Dask setup.

```python
with Client(n_workers=20, memory_limit="4GB", threads_per_worker=1) as client:
    display(client)
    df = dp1_eclipsing.compute()
df.shape
```

```python
df[["Name_vsx", "Type_vsx"]]
```
